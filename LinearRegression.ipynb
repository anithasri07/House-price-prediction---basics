```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices.csv")
```

```python
df
```

```python
plt.scatter(df.area, df.price, color= 'red', marker= '+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
135.78767123*3300+180616.43835616432
# y=mx+b
```

```python
reg.predict([[5000]])
```

```python
d = pd.read_csv("areas.csv")
```

```python
d.head()
```

```python
p = reg.predict(d)
# making the model learn from d and predict prices and store it in p
```

```python
d['prices'] = p
```

```python
d
```

```python
d.prices= d['prices'].round(2)
```

```python
d
```

```python
d.to_csv('Prediction.csv')
```
